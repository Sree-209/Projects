```python
# Load dataset
df = pd.read_csv("dataset/sign_mnist_train.csv")  # Replace with your actual path

# Get last 5 samples
last_5 = df.tail(5)

# Prepare figure
plt.figure(figsize=(15, 3))  # Wider because it's 5 images in one row

# Loop through and plot each image
for i, (idx, row) in enumerate(last_5.iterrows()):
    label = row['label']
    pixels = row.drop('label').values.astype(np.uint8)
    
    pixel_count = len(pixels)
    side_length = int(math.sqrt(pixel_count))

    if side_length * side_length != pixel_count:
        print(f"Skipping row {idx} â€“ pixel count {pixel_count} is not a perfect square.")
        continue

    image = pixels.reshape(side_length, side_length)

    # Plot image in a grid
    plt.subplot(1, 5, i + 1)
    plt.imshow(image, cmap='gray', vmin=0, vmax=255)
    plt.title(f"Label: {label}")
    plt.axis('off')

plt.tight_layout()
plt.show()

```
